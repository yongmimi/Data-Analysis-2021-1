## 데이터 시각화

### Matplotib 시작하기

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Line Plot

```python
plt.plot([1, 4, 9, 16]) # x를 생략하면 [0, 1, 2,3]
plt.show()
```

```python
plt.plot([10, 20, 30, 40],[1, 4, 9, 16]) # x,y 값 둘다 입력 
plt.title('Line Plot') # 타이틀 입력 
plt.grid() # 격자 배경
plt.show()
```

```python
import numpy as np 
x = np.arange(-4, 4.01, 0.01)
y = np.cos(x)
plt.plot(x, y)
plt.title('cos graph')
plt.grid()
plt.show()
```
