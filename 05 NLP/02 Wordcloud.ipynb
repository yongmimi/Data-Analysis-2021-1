# 워드 클라우드 

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

### Alice in Wonderland

```python
text = open('data/Alice.txt').read()
stopwords = set(STOPWORDS)
stopwords.add('said')
```

```python
len(STOPWORDS), type(STOPWORDS)
```

```python
wc = WordCloud(background_color="white", max_words=2000, stopwords=stopwords)
wc = wc.generate(text)
```

```python
type(wc.words_)
```

```python
keys = list(wc.words_.keys())
values = list(wc.words_.values())
for i in range(10):
    print(f'{keys[i]} : {values[i]}')
```

```python
plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

 - 배경 그림에 표시하기 

```python
mask = np.array(Image.open('data/Alice_mask.png'))
plt.figure(figsize=(8, 8))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear' )
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color="white", max_words=1000, stopwords=stopwords, mask=mask)
wc = wc.generate(text)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Star wars - A new Hope

```python
with open('data/A_new_hope.txt') as fd:
    text = fd.read()
text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')

mask = np.array(Image.open('data/Stormtrooper_mask.png'))
```

```python
stopwords = set(STOPWORDS)
stopwords.add('int')
stopwords.add('ext')
```

```python
wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords,  margin=10, random_state=1).generate(text)
defalut_colors = wc.to_array()
```

```python
import random

def grey_color(word, font_size, position, orientation,
               random_state=None, **kwargs):
    return 'hsl(0, 0%%, %d%%)' % random.randint(60,100)     # hue, saturation, lightness
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```
